Ова свеска прати садржај лекције о линеарној регресији. У њој можеш да се опробаш у решавању задатка одређивања цене некретнина на основу квадратура.

Учитај прво библиотеке које ће нам требати у даљем раду.

```python
import numpy as np
from matplotlib import pyplot as plt
np.random.seed(37)
```

Сада изврши следећу ћелију да би генерисао скуп података о некретнинама. Информације о квадратурама су представљене променљивом `x` док су информације о њиховим ценама представљене променљивом `y`.

```python
x = np.array([43, 25, 66, 80, 105, 70, 40, 85, 84, 102])
y = np.array([60, 32.1, 88.4, 111.4, 120.32, 72.1, 46.3, 90.1, 99.6, 139.2])
```

Изврши и следећу ћелију да би добио графички приказ тачака у скупу.

```python
plt.title('Цене некретнина')
plt.xlabel('Квадратура')
plt.ylabel('Цена')
plt.grid('on')
plt.scatter(x, y)
plt.show()
```

У следећој ћелији се налази функција која ће исцртавати праву коју одабереш у вежби. Одмах испод ње је и функција која исцртава праву коју одабереш и линијице које означавају твоје грешке. Можда ће ти бити од користи приликом тражења најбољих вредности параметара.

```python
def nacrtaj_grafik(beta_0, beta_1):
  plt.title('Цене некретнина')
  plt.xlabel('Квадратура')
  plt.ylabel('Цена')
  plt.grid('on')
  plt.scatter(x, y)

  model_x = np.linspace(20, 120, 100)
  model_y = beta_0 + beta_1*model_x

  plt.plot(model_x, model_y, color='red')

  plt.show()
```

```python
def nacrtaj_grafik_sa_greskama(beta_0, beta_1):
  plt.title('Цене некретнина')
  plt.xlabel('Квадратура')
  plt.ylabel('Цена')
  plt.grid('on')
  plt.scatter(x, y)

  model_x = np.linspace(20, 120, 100)
  model_y = beta_0 + beta_1*model_x

  plt.plot(model_x, model_y, color='red')

  for i in range(0, 10):
    prediction = beta_0 + beta_1*x[i]
    if prediction > y[i]:
      ymin = y[i]
      ymax = prediction
    else:
      ymin = prediction
      ymax = y[i]

    plt.vlines(x=x[i], ymin=ymin, ymax=ymax, colors='blue', linestyles='dotted')

  plt.show()
```

Опробај се сада са избором параметара $\beta_0$ и $\beta_1$: њихову вредност одабери на слајдеру са десне стране, а потом изврши ћелију да видиш резултат. Треба да одабереш пар параметара који по твом мишљењу најбоље одговарају расположивим подацима. Почетне вредности параметара су $\beta_0=3.84$ и $\beta_1=0.62$

```python
beta_0 = 3.84 #@param {type:"slider", min:-1, max:15, step:0.01}
beta_1 = 0.62 #@param {type:"slider", min:-2, max:4, step:0.01}
print("Odabrani model: y = {beta_0} + {beta_1}*x".format(beta_0=beta_0, beta_1=beta_1))
nacrtaj_grafik(beta_0, beta_1)
```

У доњу ћелију треба да унесеш вредности параметара $\beta_0$ и $\beta_1$ за које си се одлучио, а потом и да извршиш ћелију са позивом функције `nacrtaj_grafik_sa_greskama` која ће ти приказати грешке. Ако мислиш да можеш да направиш боље изборе, врати се на претходни корак.

Функција грешке која се придружује моделу линеарне регресије зове се `срењеквадратна грешка` (енгл.mean squared error). Следећа функција ти може помоћи да израчунаш средњеквадратну грешку за свој избор параметара.

```python
def izracunaj_srednjekvadratnu_gresku_modela (beta_0, beta_1, x, y):
  y_predikcijе  = beta_0 + beta_1*x
  pojedinacne_greske = y - y_predikcijе
  return np.average(pojedinacne_greske**2)
```

```python
izracunaj_srednjekvadratnu_gresku_modela(beta_0_odabrano, beta_1_odabrano, x, y)
```
