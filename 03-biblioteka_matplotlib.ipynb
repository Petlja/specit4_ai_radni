Matplotlib је библиотека језика Python која се користи за 2D и 3D графику. Графички прикази су нам јако корисни у раду са подацима јер нам омогућавају да боље разумемо податке, као и да финије испратимо нека понашања алгоритама.

Да би могли да користимо библиотеку matplotlib, прво морамо да је учитамо. То ћемо урадити следећом наредбом.

```python
import matplotlib.pyplot as plt
```

Учитаћемо и библиотеку `numpy` пошто ће нам требати за креирање скупова тачака за приказ.

```python
import numpy as np
```

## Цртање графика функције

У овом примеру нацртаћемо график функције $y = sin(x)$ на интервалу $[0, 10]$.

Прво ћемо формирати густу мрежу тачака дуж x осе у интервалу [0, 10] које су равномерно распоређене (такву мрежу зовемо еквидистантном мрежом) како би наш график фино изгледао. То ћемо урадити позивом функције `linspace` библиотеке numpy. Ова функција очекује као аргументе крајеве интервала 0 и 10 и број подеоних тачака - у нашем случају то може да буде број 100.

```python
x = np.linspace(0, 10, 100)
```

Потом ћемо израчунати вредност синусне функције за зваку од ових тачака позивом функције `sin(x)`. Њене вредности ћемо сачувати у променљивој `y` - и то ће сада бити један низ од 100 тачака јер се функција примењује над сваким елементом низа `x`.


```python
y = np.sin(x)
```

Сада ћемо нацртати график па објаснити шта тачно који део значи.

```python
# podesavanje naslova grafika i obelezja osa
plt.title('Grafik funkcije y=sin(x)')
plt.xlabel('x')
plt.ylabel('y')

# iscrtavanje grafika
plt.plot(x, y)

# prikaz grafika
plt.show()
```

Наслов графика постављамо позивом функције `title`, а обележја оса позивима функција `xlabel` и `ylabel`. Све ове функције су дефинисане на нивоу панела за цртање `plt`. Сам график се исцртава позивом функције `plot` и задавањем вредност `x`  и `y` координата за цртање (зато смо тим именима и назвали полазни скуп тачака и вредности синусне функције). График се приказује позивом функције `show`.


## Цртање тачкастог графикона

```python
np.random.seed(7)
```

Тачкасте графиконе често користимо за увид у просторни распоред података. У примеру који следи креираћемо десет парова тачака са целобројним вредностима координата из интервала  [0, 20] и приказати их у виду тачкастог графикона.



Прво ћемо генерисати парове тачака и то тако што ћемо засебно генерисати вредности за x и y. У томе ће нам помоћи функција `randomint` библиотеке numpy чији аргументи `low`, `heigh` и `size` омогућавају контролу доње и горње границе интервала, као и контролу броја тачака.

```python
# nizovi duzine 10 sa proizvoljnim elementima iz intervala [0, 20]
x = np.random.randint(low=0, high=20, size=10)
y = np.random.randint(low=0, high=20, size=10)
```

Сам тачкасти графикон се креира позивом функције `scatter` панела за цртање `plt`. Приликом позива овој функцији се задају вредности координата тачака, у нашем случају `x` и `y`. Додатно се може подесити боја тачака аргументом `color`, као и изменити сам симбол за приказ аргументом `marker`. На графику су уместо подразумеваних кружића црне боје коришћени зелени троугилићи усмерени на доле. График се приказује, као и у претходном примеру, позивом функције `show`.

```python
# generisanje tackastog grafikona
plt.scatter(x, y, color='green', marker='v')

# prikaz grafika
plt.show()
```

Можемо да приметимо да смо у овом примеру на нивоу библиотеке numpy и њеног пакета `random` подесили генератор случајних бројева (такозвано својство `seed`) на вредност 7. То ће нам омогућити да сваки пут када покренемо овај код добијемо исти распоред тачака. Ово својство нам је важно због могућности поновног покретања експеримената.
